```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1=pd.read_csv(r"match_data\ban-vs-ind.csv")
df2=pd.read_csv("match_data\ind-vs-pak-super-fours.csv")
df3=pd.read_csv("match_data\ind-vs-pak.csv")
df4=pd.read_csv("match_data\ind-vs-sl.csv")
df5=pd.read_csv("match_data\sl-vs-ind.csv")
df=pd.concat([df1,df2,df3,df4,df5])
df.drop(columns=["ball_speed","Unnamed: 0","commentary"],axis=1,inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

```python
df.batter.value_counts()
```

```python
df_Ishan_Kishan_batting=df[df.batter=="Ishan Kishan"]

import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Create a 3x1 subplot grid
fig, axes = plt.subplots(1,3, figsize=(16, 5))

# Plot on the first subplot (axes[0])
axes[0].scatter(df_Ishan_Kishan_batting.shot,df_Ishan_Kishan_batting.runs, cmap='viridis', alpha=0.5)

# Plot on the second subplot (axes[1])
axes[1].scatter(df_Ishan_Kishan_batting.ball_swing,df_Ishan_Kishan_batting.runs, cmap='viridis', alpha=0.5)

# Plot on the third subplot (axes[2])
axes[2].scatter(df_Ishan_Kishan_batting.ball_length,df_Ishan_Kishan_batting.runs, cmap='viridis', alpha=0.5)

# Add labels and titles if needed
axes[2].set_xlabel('Ball Length')
axes[1].set_xlabel('Ball Swing')
axes[0].set_xlabel('Shot')
axes[0].set_ylabel('runs')
axes[1].set_ylabel('runs')
axes[2].set_ylabel('runs')
fig.suptitle('Batter(Ishan Kishan) Score Stats')

# Adjust subplot spacing
plt.tight_layout()
plt.legend(title="Point Density Propotional to Count",loc='upper right',bbox_to_anchor=(1.05, 1.2))
# Show the plot
plt.show()
```

```python
df_Ishan_Kishan_batting=df[df.batter=="Ishan Kishan"]

import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Create a 3x1 subplot grid
fig, axes = plt.subplots(1,3, figsize=(16, 5))

# Plot on the first subplot (axes[0])
axes[0].hist(df_Ishan_Kishan_batting.ball_length,alpha=0.8,edgecolor="black",density=True)

# Plot on the second subplot (axes[1])
axes[1].hist(df_Ishan_Kishan_batting.ball_swing,alpha=0.8,edgecolor="black",density=True)

# Plot on the third subplot (axes[2])
axes[2].hist(df_Ishan_Kishan_batting.shot,alpha=0.8,edgecolor="black",density=True)

# Add labels and titles if needed
axes[0].set_xlabel('Ball Length')
axes[1].set_xlabel('Ball Swing')
axes[2].set_xlabel('Shot')
axes[0].set_ylabel('frequency')
axes[1].set_ylabel('frequency')
axes[2].set_ylabel('frequency')
fig.suptitle('Batter(Ishan Kishan) Wicket Stats')

# Adjust subplot spacing
plt.tight_layout()
plt.xticks(rotation=45)
# Show the plot
plt.show()
```

```python
df_Kuldeep_Yadav_bowling=df[df.baller=="Kuldeep Yadav"]

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Plot on the first subplot (axes[0]) - Scatter Plot
axes[0].scatter(df_Kuldeep_Yadav_bowling['batter'], df_Kuldeep_Yadav_bowling['runs'], cmap='viridis', alpha=0.1)
axes[0].set_ylabel('Runs')
axes[0].set_xlabel('Batter')
axes[0].set_title('Scatter Plot: Runs vs. Batter')
axes[0].grid(True)
axes[0].tick_params(axis='x', rotation=45)

# Plot on the second subplot (axes[1]) - Histogram
df_Kuldeep_Yadav_bowling=df[df.wicket==1]

axes[1].hist(df_Kuldeep_Yadav_bowling['batter'], alpha=0.8, edgecolor="black", density=True)
axes[1].set_xlabel('Batter')
axes[1].set_ylabel('Wickets Frequency')
axes[1].set_title('Histogram: Wickets')
axes[1].grid(True)

# Rotate the x-axis tick labels by 45 degrees
plt.xticks(rotation=45)

# Add a common title for the entire figure
fig.suptitle('Performance Stats for Baller(Kuldeep Yadav)', fontsize=16)

# Adjust subplot spacing
plt.tight_layout()

# Show the plot
plt.show()
```
